```python
x = np.linspace(-5, 5, 400) #array di valory equispaziati tra -5 e 5
```

```python
y = np.exp(x)
plt.plot(x, y)
```

```python
x = np.linspace(0, 4*np.pi, 400)  #np.pi vuol dire PIgreco
y = np.sin(x) 
plt.plot(x,y)
```

```python
y[y > 0.5] = 0.5 #indicizzazione boolean, evidenziamo i punti maggiori di 0,5 e solo per quelli assegnamo 0,5 per questo escono le parti piatte sopra perche superano 0.5
plt.plot(x,y)
```

```python
x = np.linspace(-20*np.pi, 20*np.pi, 10000)
y = np.sin(x) / x
plt.plot(x, y)
```
