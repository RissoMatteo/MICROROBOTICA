```python
def createNgramDictionary(text, N):
    diz_Ngramma = {}
    for posizione in range(0, len(text) - N):
        bi = text[posizione: posizione + N]
        if bi in diz_Ngramma:
            diz_Ngramma[bi] += 1
        else:
            diz_Ngramma[bi] = 1
    return diz_Ngramma
```

```python
list_righe = []
with open("OSCAR_WILDE_Il_fantasma_di_Canterville.txt", 'r') as file:
    list_righe = file.readlines()
```

```python
nrighe = 0
list_righe = [riga for riga in list_righe if len(riga)>1]
nrighe = len(list_righe)
print(f"Le righe sono: {nrighe}")
```

```python
num_caratteri = 0;
for riga in list_righe:
    num_caratteri += len(riga)
print(f"I caratteri sono: {num_caratteri}")
```

```python
riga_lung_max = ""
for riga in list_righe:
    if len(riga) > len(riga_lung_max):
        riga_lung_max = riga;
print(f"La riga più lunga è: {riga_lung_max}")
```

```python
dizionario = {}
for riga in list_righe:
    for carattere in riga:
        if carattere not in dizionario:
            dizionario[carattere] = 1;
        else:
            dizionario[carattere] += 1;
```
