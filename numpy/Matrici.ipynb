```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dim = 8
chessboard = np.zeros((dim, dim), dtype = np.int32)
#slicing assegnazione
chessboard[::2, ::2] = 1 #seleziona le righe saltanso di 2 in due, RIGHE DISPARI
chessboard[1::2, 1::2] = 1 #quando sei su una riga salta le celle scorrendo per le colonne a 2 a 2, RIGHE PARI
plt.imshow(chessboard, cmap = "Greys")
```

```python
dim = 10
array2d = np.zeros((dim, dim), dtype = np.float32)
array2d[:] = np.linspace(0.,1.,dim)  #genera valori equidistanti tra 0 e 1 di dim campi, [:] = tutte le righe
plt.imshow(array2d, cmap = "magma")
```

```python
dim = 10
array2d = np.zeros((dim, dim), dtype = np.float32)
array2d[:] = np.linspace(0.,1.,dim)  #genera valori equidistanti tra 0 e 1 di dim campi, [:] = tutte le righe
array2d[0] = 0 #prima riga
array2d[-1] = 0 #ultima riga
array2d[:,0] = 0 #prima colonna
array2d[:,-1] = 0 #ultima colonna
plt.imshow(array2d, cmap = "magma")
```

```python
dim = 10
bersaglio2d = np.zeros((dim, dim), dtype=np.float32)

for i in range(0, dim +1):
    bersaglio2d[i:-i, i:-i] = i # 

plt.imshow(bersaglio2d, cmap="magma")
```

```python
dim = 10
noise = np.random.uniform(0.,1.,(dim,dim)) #matrice di numeri a caso di dim rige e dim colonne
plt.imshow(noise, cmap = "jet")
```

```python
red_pixels = noise > 0.8 #array di boolean, considerando la scala da 0 a 1
red_pixels
noise[red_pixels] = 0
plt.imshow(noise, cmap = "jet", vmax = 1, vmin = 0) #spariscono i rossi
```
