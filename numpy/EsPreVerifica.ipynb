Caricare un immagine e creare l effetto del selfie della fotocamera: effetto specchio

```python
import numpy as np
import matplotlib.pyplot as plt

img = plt.imread("imgNetflix.jpeg")

# Crea l'effetto specchio
mirror_img = img[:,::-1,:]  #righe, colonne, rgb  -> : = tutte le righe, ::-1 = inverto tutte le colonne, : = prendo tutti i canali RGB per non cambiasre i colori

plt.imshow(mirror_img)
```

```python
#Utilizza numpy per simulare i 27 * 10 voti interi, dopo calcola quanti alunni hanno un voto medio > 6.
#Non usare For, usa lo slicing e fallo in modo semplice. 

import numpy as np

voti = np.random.randint(1, 11,(27, 10))
voti_medi = np.mean(voti, axis=1) 
num_alunni = np.sum(voti_medi > 6)

print("Numero di alunni con voto medio > 6:", num_alunni)
```
