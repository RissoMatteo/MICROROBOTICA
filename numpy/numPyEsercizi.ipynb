**La FUNZIONE np.roll()**

```python
array = np.arange(1, 6)
array
```

```python
np.roll(array , 1) #schifta di 1 o il valore che gli do veso destra o sinistra a seconda se il valore è + o -
```

**Dato un array di numeri, usare numpy  per determinare se l array è monotono crescente**

```python
array = np.array([1,2,5,6,7,10])
array
```

```python
np.roll(array, -1) #scorro di uno 
```

```python
(np.roll(array, -1) > array)[:-1] #faccio uscire i valori maggiori del successivo se si = ture
```

```python
np.count_nonzero((np.roll(array, -1) > array)[:-1]) #conto i true
```

```python
if np.count_nonzero((np.roll(array, -1) > array)[:-1]) == (len(array)-1):
    print("Array monotono")
else:
    print("Array non monotono")
```

```python
immagine = plt.imread("./CompitoImmgini.jpg").copy()
colonne, larghezza, _ = immagine.shape
red = immagine[:,:,0]
green = immagine[:,:,1]
blue = immagine[:,:,2]
immagine[:,:,0] = np.roll(red, 0)
immagine[:,:,1] = np.roll(green, colonne // 3)
immagine[:,:,2] = np.roll(blue, colonne // 3 * 2)
plt.imshow(immagine)
```

```python
#controllo dei massimi e minimi locali  
array = np.array([1,2,3,20,5,6,8,9,10])
array
```

```python

array_shift_right = np.roll(array, 1)
array_shift_left = np.roll(array, -1)
local_max = (array > array_shift_left) & (array > array_shift_right)
local_max
array[local_max]
```
