**Crea una cornice blu sull immagine con spessore 10px**

```python
image = plt.imread("./prova.jpeg").copy()
image[:10, :, 2] = 255 #prime 10 righe in alto
image[:10, :, :2] = 0

image[:, :10, 2] = 255 #bordo snistro
image[:, :10, :2] = 0

image[:, -10:, 2] = 255 #bordo destro
image[:, -10:, :2] = 0

image[-10:, :, 2] = 255 #ultime 10 righe in basso
image[-10:, :, :2] = 0

plt.imshow(image)
```
