```python
import numpy as np
import matplotlib.pyplot as plt
```

**Caricare una immagine e creare una scacchiera sul suo canale rosso, nella sola metà sinistra dell’immagine**

```python
immagine = plt.imread("./CompitoImmgini.jpg").copy()
_ ,larghezza, _ = immagine.shape #cerlo il valore larghezza dell'immagine per poi divederla a metà
meta_larghezza = larghezza // 2
immagine[::2, :meta_larghezza:2, 0] = 255  # celle pari con il massimo rosso
immagine[1::2, 1:meta_larghezza:2, 0] = 255  # celle dispari con il massimo rosso

plt.imshow(immagine)
```

**caricare una immagine e creare l’immagine in scala di grigio ottenuta sommando tutti i canali dell’immagine originale.**

```python
immagine = plt.imread("./CompitoImmgini.jpg").copy()
red = immagine[:,:,0]  #tiro fuori il canale rosso
green = immagine[:,:,1] #verde
blue = immagine[:,:,2] #blu
immagine_grigia = red + green + blue #sommo i canali

plt.imshow(immagine_grigia, cmap='Greys')
plt.show()
```

**caricare una immagine e permutare i suoi canali: R diventa G, G diventa B e B diventa R**

```python
immagine = plt.imread("./CompitoImmgini.jpg").copy()
red = immagine[:,:,0]  #tiro fuori il canale rosso
green = immagine[:,:,1] #verde
blue = immagine[:,:,2] #blu

immagine[:,:,0] = green #scambio i canali
immagine[:,:,1] = blue
immagine[:,:,2] = red

plt.imshow(immagine)
```
