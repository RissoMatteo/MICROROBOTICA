```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#grafico di una curv gaussiana
x = np.linspace(-5, 5, 1000)
y = np.exp(-x**2)
x_tratteggiato = np.linspace(-6,6,1000)
y_tratteggiato = np.exp(-x_tratteggiato**2)
fig, ax = plt.subplots(figsize=(6,4))
ax.plot(x_tratteggiato, y_tratteggiato, linestyle = "--")
fig.set_facecolor("white")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.grid()
ax.plot(x, y, linewidth = 2)
plt.show()
```

```python
#grafici a dispersione
n_punti = 100
x = np.random.uniform(0.,1., n_punti)
y = np.random.uniform(0.,1., n_punti)
x_1 = np.random.uniform(0.,1., n_punti)
y_1 = np.random.uniform(0.,1., n_punti)

fig, ax = plt.subplots(figsize = (5,5))
ax.plot(x,y, linestyle = " ", marker = "*", color = "white", markersize = 6) #nessuna linea ma come marker usa un pallino
ax.plot(x_1,y_1, linestyle = " ", marker = "+", color = "yellow", markersize = 10) #nessuna linea ma come marker usa un pallino

ax.set_facecolor("blue")
plt.show()
```
