```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#rappresentiamo dei dati:
x = np.linspace(-5.,5.,1000) #1000 valori equispaziati tra -5, 5
y = x ** 2 #  ** vuol dire alla seconda
fig, ax = plt.subplots(figsize=(4,2))   #ax sono gli assi, fig è il mio foglio bianco
y1 = x ** 2
y2 = 1 / x
ax.plot(x, y1, label = "y1", color = "blue", alpha = 0.5, linewidth = 10, linestyle = "--") #l posto di "y1" si puo anche mettere -> $y_1 = x^2$, alpha = trasparenza, linewidth = spessore, linestyle = tipologia di linea 
ax.plot(x, y2, label = "y2", linestyle = "-.")
ax.set(xlabel="x") #vado a dare dei nomi, delle etichette all asse x
ax.set(xlabel="y")
ax.set_title("Questo è il titolo del grafico", fontsize=10)
ax.grid()
ax.set_ylim([-10,10])
ax.legend() #legenda, data dai label del ax.plot(x, y1, label = "y1")
plt.show()
```
